```python
import arty
from matplotlib import pyplot as plt
plt.rcParams["figure.figsize"] = (10, 10)
```

```python
img = (plt.imread('images/img.png') * 255).astype(int)
plt.imshow(img)
plt.show()
```

```python
dx, dy = arty.edge.gradientX(img), arty.edge.gradientY(img)
edges = arty.edge.gradient(img)
angles = arty.angle.gradient_truth(dx, dy)
lines = arty.filter.lines(img, angles, edges)

plt.imshow(lines)
plt.show()
```

```python
blur = arty.filter.blur.gauss(img)
dx, dy = arty.edge.gradientX(blur), arty.edge.gradientY(blur)
edges = arty.edge.gradient(blur)
angles = arty.angle.gradient_truth(dx, dy)
lines_blur = arty.filter.lines(blur, angles, edges)

plt.imshow(lines_blur)
plt.show()
```

```python
noised = arty.filter.noise(blur)
lines_noise = arty.filter.lines(noised, angles, edges)

plt.imshow(lines_noise)
plt.show()
```
